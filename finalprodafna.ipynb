## Foreign investment rose till 2001-2002, then has seen a gradual drop with some inconsistent rise. So far, I have only been using barplots as an indicatior, Lets work with some boxplots. I will be using boxplots to compare performance over a period of decades

```python
def function(x):
    if x>=1960 and x<1970:
        return '1960-1970'
    elif x >=1970 and x<1980:
        return '1970-1980'
    elif x>=1980 and x<1990:
        return '1980-1990'
    elif x>=1990 and x<2000:
        return '1990-2000'
    elif x>=2000 and x<2010:
        return '2000-2010'
    elif x>=2010 and x<2020:
        return '2010-2019'
    
```

```python
France_t['years'] = pd.to_numeric(France_t['years'])
```

```python
France_t['decades'] = France_t['years'].apply(lambda x:function(x))
```

```python

plt.figure(figsize = (8,8))
sns.boxplot(France_t['decades'],France_t['Adjusted net national income (annual % growth)'])

```

```python
plt.figure(figsize = (12,8))
sns.barplot(France_t['decades'],France_t['GDP per capita (current US$)'])
```

```python
plt.figure(figsize = (14,6))
plt.subplot(1,2,1)
sns.boxplot(France_t['decades'],France_t['GDP per capita growth (annual %)'])
```

```python

plt.figure(figsize = (12,8))
sns.boxplot(France_t['decades'],France_t['Foreign direct investment, net inflows (% of GDP)'])
```

```python
plt.figure(figsize = (14,6))
plt.subplot(1,2,1)
sns.boxplot(France_t['decades'],France_t['GDP per capita growth (annual %)'])
```

## The boxplots convey the same results as the barplots but over the course of decades. Overall France GDP and annual net income has grown but it's rate of growth has declined. This can be due to an increase in the aging population and a decline in young people as seen 

```python
plt.figure(figsize = (14,6))
plt.subplot(1,2,1)
sns.boxplot(France_t['decades'],France_t['Adjusted net national income (constant 2010 US$)'])
plt.subplot(1,2,2)
sns.boxplot(France_t['decades'],France_t['Foreign direct investment, net inflows (% of GDP)'])
```
